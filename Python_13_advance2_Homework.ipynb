```python
#Задание 1
import pandas as pd
df = pd.read_csv('ratings.csv')
df = df.groupby('userId').agg({'movieId':'count', 'rating':'count', 'timestamp':['min','max']})
df = df[df['rating']['count'] > 100]
df['ts_diff'] = df['timestamp']['max'] - df['timestamp']['min']
df
```

```python
#Задание 2
rzd = pd.DataFrame(
    {
        'client_id': [111, 112, 113, 114, 115],
        'rzd_revenue': [1093, 2810, 10283, 5774, 981]
    }
)
rzd
```

```python
auto = pd.DataFrame(
    {
        'client_id': [113, 114, 115, 116, 117],
        'auto_revenue': [57483, 83, 912, 4834, 98]
    }
)
auto
```

```python
air = pd.DataFrame(
    {
        'client_id': [115, 116, 117, 118],
        'air_revenue': [81, 4, 13, 173]
    }
)
air
```

```python
client_base = pd.DataFrame(
    {
        'client_id': [111, 112, 113, 114, 115, 116, 117, 118],
        'address': ['Комсомольская 4', 'Энтузиастов 8а', 'Левобережная 1а', 'Мира 14', 'ЗЖБИиДК 1', 
                    'Строителей 18', 'Панфиловская 33', 'Мастеркова 4']
    }
)
client_base
```

```python
# сформировать таблицу с тремя типами выручки для каждого client_id без указания адреса клиента
data = (rzd.merge(auto, how = 'outer')
        .merge(air, how = 'outer')
       )
data
```

```python
# аналогичную таблицу по типам выручки с указанием адреса клиента
data = data.merge(client_base, how = 'outer')
data
```
